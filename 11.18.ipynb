```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
#查看数据集的前几行
print(iris_df.head())

#查看数据集的后几行
print(iris_df.tail(5))
print("数据集的维度：", iris_df.shape)
#查看数据集的基本信息，包括每列的数据类型和缺失值
iris_df.info()
#查看数据集的描述统计信息
print(iris_df.describe())
```

```python
iris_df.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot (data=iris_df)
plt.tight_layout()
plt.show ()
plt.savefig('boxplot_features.png')
```

```python
correlation_matrix = iris_df.drop(columns="species").corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.show()
```

```python
sns.pairplot(iris_df, hue="species", height=3)
```
